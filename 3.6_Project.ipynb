```python
from datetime import datetime
import pandas as pd
from IPython.display import display, Markdown
import matplotlib.pyplot as plt
# Load the dataset into a new DataFrame
df_cleaned =pd.read_excel('C:\\Files\\Excel\\Predictive_PythonR\\Happiness_Development_Index_Bradley.xlsx')

# drop empty columns
df_cleaned = df_cleaned.loc[:, ~ df_cleaned.columns.str.contains("^Unnamed")]
# remove extra spaces from columns
df_cleaned.columns = df_cleaned.columns.str.strip()

import pandas as pd #import pandas library
current_datetime = datetime.now().strftime("%Y-%m-%d")
student_id = "Jerbra1624"
display(Markdown(f"### {student_id} – {current_datetime}")) #display Header Text
```

```python
df_cleaned.columns
```

```python
# Histogram of HDI Scores 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df_cleaned["Human Development Index (HDI)"] = pd.to_numeric(
    df_cleaned["Human Development Index (HDI)"],
    errors="coerce"
)


plt.figure(figsize=(10,6))
plt.hist(df_cleaned["Human Development Index (HDI)"].dropna(), bins=10,color='skyblue', edgecolor='black' )
plt.title(f'Distribution of HDI Scores Across Countries: {student_id} - {current_datetime}')
plt.xlabel('HDI Score')
plt.ylabel('Number of Countries')
plt.show()




```

```python
# Bar chart of Top 10 coutries by highest HDI 
top10 = df_cleaned.sort_values('Human Development Index (HDI)', ascending=False).head(10)


plt.figure()
plt.barh(top10['Country'], top10['Human Development Index (HDI)'])
plt.title(f'Top 10 Countries by HDI Score: {student_id} - {current_datetime}')
plt.xlabel("HDI Score")
plt.ylabel("Country")
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.figure()
plt.scatter(df_cleaned['Gross national income (GNI) per capita'], df_cleaned['Human Development Index (HDI)'])
plt.title('HDI vs GNI per Capita:')
plt.xlabel('Gross national income (GNI) per capita')
plt.ylabel('Human Development Index (HDI)')
plt.show()
```

```python
selected_columns = ['Human Development Index (HDI)', 'Life expectancy at birth', 'Expected years of schooling', 'Mean years of schooling', 'Gross national income (GNI) per capita']
correlation_matrix = df_cleaned[selected_columns].corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=" .2f")
plt.title(f'Heatmap of correlations: {student_id} - {current_datetime}')
plt.show()
```
